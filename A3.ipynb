## Did more earthquakes happen on weekends or weekdays? 

```python
df = pd.read_csv("新建文件夹\cleaned_earthquakes_5.5_1960-2023.csv")
df['time'] = pd.to_datetime(df['time'])
df['day_of_week'] = df['time'].dt.day_name()
day_counts = df['day_of_week'].value_counts()
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_counts = day_counts.reindex(days_order)
fig, ax1 = plt.subplots(figsize=(10, 6))
sns.barplot(x=day_counts.index, y=day_counts.values, order=days_order, palette='viridis', ax=ax1)
ax1.set_title('Earthquakes Occurrences by Day of the Week')
ax1.set_xlabel('Day of the Week')
ax1.set_ylabel('Number of Earthquakes')
ax2 = ax1.twinx()
day_counts.plot(kind='line', marker='o', color='red', ax=ax2)
ax2.set_ylabel('Number of Earthquakes (Line Graph)')
plt.show()
```

## Comment on changes in the data with time: a) earthquakes per year, b) type of earthquakes, and c) magnitude resolution. Include a figure

#### a) Earthquakes per year:

```python
earthquake_data = pd.read_csv("新建文件夹\cleaned_earthquakes_5.5_1960-2023.csv")
earthquake_data['year'] = pd.to_datetime(earthquake_data['time']).dt.year
earthquakes_per_year = earthquake_data['year'].value_counts().sort_index()
plt.figure(figsize=(14, 8))
sns.lineplot(x=earthquakes_per_year.index, y=earthquakes_per_year.values, marker='o')
plt.title('Earthquakes Per Year (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Number of Earthquakes')
plt.grid(True)
for i, txt in enumerate(earthquakes_per_year.values):
    plt.text(earthquakes_per_year.index[i], txt, str(earthquakes_per_year.index[i]))
plt.show()
```

```python
df = pd.read_csv("cleaned_earthquakes_5.5_1960-2023.csv")
df['year'] = pd.to_datetime(df['time']).dt.year
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
kde_plot = sns.kdeplot(df['year'], fill=True, palette="viridis", legend=False)
plt.scatter(df['year'], [0.005] * len(df), color='red', alpha=0.5, label='Earthquake Events')
plt.axhline(0.01, color='blue', linestyle='dashed', linewidth=2, label='Threshold Line')
plt.title('Distribution of Earthquakes per Year (1960-2023) with Points and Line')
plt.xlabel('Year')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### b) Type of earthquakes:

```python
df = pd.read_csv("cleaned_earthquakes_5.5_1960-2023.csv")
df['time'] = pd.to_datetime(df['time'])
df_filtered = df[(df['time'].dt.year >= 1990) & (df['time'].dt.year <= 2023)]
unique_earthquake_types = df_filtered['type'].unique()
print(unique_earthquake_types)
```

```python
df['year'] = pd.to_datetime(df['time']).dt.year
df_filtered = df[(df['year'] >= 1990) & (df['year'] <= 2023)]
earthquake_counts = df_filtered.groupby(['year', 'type']).size().unstack()
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']  # Add more colors if needed
ax = earthquake_counts.plot(kind='bar', stacked=True, color=colors, figsize=(12, 6))
ax.set_title('Distribution of Earthquake Types (1990-2023)')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Earthquakes')
ax.legend(title='Earthquake Type')
plt.show()
```
